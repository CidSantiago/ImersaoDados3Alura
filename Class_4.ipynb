## Challenge 01

To do this, i will start from the `'contagem_moa'` series previously made, and work from there:

```python
new_list = []
for name in contagem_moa.index:
    new_list.append(name.split('_')[-1])

new_contagem_moa = contagem_moa
new_contagem_moa.index = new_list
new_contagem_moa.groupby(level=0).sum().sort_values(ascending=False)[:10]
```

Noice(sic)! The first challenge is completed! We can see that the great majority of the actions are the inhibitor(9693),antagonist(3449) and agonist(2330). Even if we add the other actions withdrawing those three we didn't get close to the agonist. Look!:

```python
new_contagem_moa.groupby(level=0).sum().sort_values(ascending=False)[3:].sum()
```

So, the explaining of how i did it:
First, i created a new_list. That would consist of the new index in the `contagem_moa` series. Then, I get all the index in `contagem_moa`, splitted the string using the "_" as delimiter, and get the last part after the delimiter. As we can notice, they follow a pattern that 'target_action', so we should get only the actions of each row like this. After that, we rename the index in a `new_contagem_moa` and, using `groupby`, we group by the names and sum it all, getting the results we want! :D

## Challenge 04
---
This challenge was to study merge, join, concatenate and compare. So i will just describe each one here:

`pandas.merge()` is very similar to the database merges. We define the keys of each dataframe that we want to merge, and tell how we want to do it (Left outer join, inner join, right inner join, etc)

`pandas.join()` is just a merge that uses only the index of the dataframes. So, if we have two dataframes that the index are the "primary key", we just use join and move on.

`pandas.concatenate()` unite dataframes horizontally or vertically (depends of what you passes as the axis). Some peculariarities happens, as if you concatenate two dataframes with different columns names in the axis=0, you create two "NaN voids" because of the unmatcheds columns. But still very useful!

`pandas.DataFrame.compare()` is self-explanatory. It compares the DataFrame with another and returns the divergences.



## Challenge 05
---

This is the hardest challenge. We have to compare each value to each different configuration. So, lets start 


```python
sns.set()

composto_principal = data_combinados['composto'].value_counts().index[:5]

def plot_experiments_box_plots(dose, tempo):
    plt.figure(figsize=(12,8))
    sns.boxplot(data = data_combinados.query(f'composto in @composto_principal and dose == "{dose}" and tempo == {tempo}'),
     y='g-0', x='composto', hue='ativo_moa', order=composto_principal)
    plt.title(f'Comparison of g-0 with dose {dose} and time {tempo} in the five more frequent compounds')

dose_list = ['D1','D2']
tempo_list = [24,48,72]

for dose in dose_list:
    for tempo in tempo_list:
        plot_experiments_box_plots(dose,tempo)
```

Cool! We did it! Now lets have some insights in this graphs:

* The second compound tends to influence more on the g-0 after 48hs. We can verify this raise in both dosages.

* The third compound had little to none influence in the g-0

* The fourth compound only affects the g-0 slighly on the first 24 hs. After that, it came back to the default, based on the control group

* The fifth compound had a consistent effect on the 'g-0', independent of the time and dosage

* In this graphs that we plotted, we can't see much diffence between the effects on g-0 independently of the dosage. The big discrepancy was between the second compound on the first 24hs.
