```python
data_resultados['n_moa'] = data_resultados.drop('id', axis=1).sum(axis=1)
```

```python
data_resultados['ativo_moa'] = (data_resultados['n_moa'] != 0)
data_resultados
```

```python
data_combinados = pd.merge(data, data_resultados[['id','n_moa', 'ativo_moa']], on='id')
data_combinados
```

```python
import seaborn as sns
import matplotlib.pyplot as plt

sns.set()

composto_principal = data_combinados['composto'].value_counts().index[:5]

plt.figure(figsize=(12,8))
sns.boxplot(data = data_combinados.query('composto in @composto_principal'), y='g-0', x='composto', hue='ativo_moa')
```
