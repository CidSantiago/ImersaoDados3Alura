Lets talk about my homie, my bro, `describe()`? This command is actually one of the first things that i do in a dataset, because it tells me so much with so little effort. Lets talk about all the things that it tells! I will first tell you what i know from my head, and then, if i think it was not satisfying enough, i will search the concepts :) To start things, let see the describe in action:

```python
data['g-0'].describe()
```

I am applying `describe()` directly in just one column, because I think that is enough to us to see. Lets go line by line!

* **Count**: The quantity of values on the series. All the values, not the uniques only.

* **std**: Standard deviation of the values in the series. Basically is how the data is distributed around the mean. If this is low, indicates that the values have a tendency to stay around the mean. This is a good indicative if the data is disperse or not.

* **min**: Minimum value in the series. Self-explanatory, right?

* **25%**: Commonly know as "first quartile". What that means? Means that this is the point of cut that indicates that if we get the values from this point below, we get 25%, or one quarter, of the series. This helps us to understand better how the values are distributed.

* **50%**: Commonly know as "median", but is also "second quartile". Two names, same thing. And makes sense, right? From the previous explanation, we should know that it refers to the point of cut that determines exactly half of the values. This IS the concept of median. Statistics are so cool 😍

* **75%**: Commonly know as "third quartile". I think you already get it, right? From this point below, represents 75% of the values, bla bla bla. Lets move on.

* **max**: Maximum value in the series. This is it.
