```python
import numpy as np

corr = data.loc[:, 'g-0':'g-50'].corr()

# Generate a mask for the upper triangle
mask = np.triu(np.ones_like(corr, dtype=bool))

# Set up the matplotlib figure
f, ax = plt.subplots(figsize=(11, 9))

# Generate a custom diverging colormap
cmap = sns.diverging_palette(230, 20, as_cmap=True)

# Draw the heatmap with the mask and correct aspect ratio
sns.heatmap(corr, mask=mask, cmap=cmap, center=0,
            square=True, linewidths=.5, cbar_kws={"shrink": .5})
```

*Here we go, **Challenge 05**! To do this i can just get 50 Gs and 50 Cs and plot the correlation matrix. Lets do it:*

P.S: Challenge 05: Calculate and analyse the correlation between G and C. Reflect on the biologics effects.

```python
#Challenge 05
#Personal

G_s = [f'g-{i}' for i in range(51)]

C_s = [f'c-{i}' for i in range(51)]

gs_and_cs = data.loc[:, G_s + C_s]

corr_GsCs = gs_and_cs.corr().loc['c-0':'c-50','g-0':'g-50']

# Generate a mask for the upper triangle
mask = np.triu(np.ones_like(corr_GsCs, dtype=bool))

# Set up the matplotlib figure
f, ax = plt.subplots(figsize=(11, 9))

# Generate a custom diverging colormap
cmap = sns.diverging_palette(230, 20, as_cmap=True)

# Draw the heatmap with the mask and correct aspect ratiocorr
sns.heatmap(corr_GsCs, mask=mask, cmap=cmap, center=0,
            square=True, linewidths=.5, cbar_kws={"shrink": .5})
```

*What i did here was to create a list with the two ranges of Gs and Cs that i wanted, and concatenate then in the `.loc`. After that, i called the `.corr()` method in all of it. After that, i sliced the result to get only the correlation between Cs and Gs. I wasted a little bit more resources that needed calling `.corr()` in all the Gs and Cs selected, but performance at the moment is not the focus! I will plot in Plotly too because i love plotly ‚ù§*

*It is important to notice that some genes(G) has high correlation with some celular types(C). What that problem means is that these specific genes probably are related to the health of the Cells. So they raise and decrease simmilarly, implying high correlation.*

*For the **Challenge 06**, i looked up the documentation of the functions that were used, and i will describe line by line what happened:*

` mask = np.triu(np.ones_like(corr_GsCs, dtype=bool)) `
*The first thing it does is create a matrix of 'True' values in the shape of the correlation matrix, using the `np.ones_like()`. After that, it gets the upper triangle of the True values matrix, returning True in the upper triangle and False in the lower triangle. With this, we get a 'mask', that will be used as a parameter of the heatmap plot*

` f, ax = plt.subplots(figsize=(11, 9)) `
*Create the matplotlib space to plot the figure. Not much going here.*

`cmap = sns.diverging_palette(230, 20, as_cmap=True)`
*Basically creates a matplotlib color palette going from blue(230) to red(10)*

`sns.heatmap(corr_GsCs, mask=mask, cmap=cmap, center=0, square=True, linewidths=.5, cbar_kws={"shrink": .5})`
*Plot the heatmap corresponding to the correlation matrix. In this function, we pass our previously defined mask, to tell what values it should plot and our color palette (mask and cmap, respectively). We also give some other parameters as `center = 0`, defining that the center of the color palette, `square=True`, to shape the plot in a way that each cell is a square, `cbar_kws={"shrink": .5}`, that shrink the size of the lateral color bar, and `linewidths=.5`, which is self-explanatory.*
